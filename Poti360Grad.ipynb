```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(20,8))
plt.axis([0, 360, 0, 4096])

# phases and increments
p1 = 0 
p2 = 2048
anstieg1 = 4096/360*2
anstieg2 = 4096/360*2

for i in range(360):
    p1=p1+anstieg1
    p2=p2+anstieg2
    if p1 > 4096: anstieg1=anstieg1*-1
    if p2 > 4096: anstieg2=anstieg2*-1
    if p1 < 0:    anstieg1=anstieg1*-1
    if p2 < 0:    anstieg2=anstieg2*-1
    # Poti ADC Values Simulation
    plt.plot(i, p1, marker = '.', markersize = 6, color="blue")
    plt.plot(i, p2, marker = 'p', markersize = 4, color="green")

        
    # gewichte berechnen  werte in der Mitte richtung 1, werte nahe minimal/maximalwert 0  
    w1 = 1-(abs(2048 - p1) / 2048 )  # Best 0, schlecht 1 , daher 1-x
    w2 = 1-(abs(2048 - p2) / 2048 ) 

    plt.plot(i, p1 * w1  , marker = '.', markersize = 3, color="purple")
    plt.plot(i, p2 * w2  , marker = '.', markersize = 3, color="red")
        
plt.show()
```
